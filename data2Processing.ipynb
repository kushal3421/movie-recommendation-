```python
#converting release_date column into date format from string
meta['release_date']=pd.to_datetime(meta['release_date'],errors='coerce')
```

```python
# creating a new dataframe with required data fields only
new_data=meta.loc[meta.year==2017,['genres','id','title','year']]
```

```python
new_data.head()
```

```python
#converting movie id into integer from object
new_data['id']=new_data['id'].astype(int)
```

```python
new_data.dtypes
```

```python
data = pd.merge(new_data,credits,on='id')
```

```python
data.head()
```

```python
#This particular statement sets the maximum width to 75 pixels per column.
pd.set_option('display.max_colwidth',75)
data.head()
```

```python
# converting string into dictionary for genre column 
import ast
data['genres']=data['genres'].map(lambda x: ast.literal_eval(x))
```

```python
# converting string into dictionary for cast and crew column 
data['cast']=data['cast'].map(lambda x: ast.literal_eval(x))
data['crew']=data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data['cast']
```

```python
def genre_list(x):
    gen=[]
    st=' '
    for i in x:
        if i.get('name')=='Science Fiction':
            scifi='Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genre_list']=data['genres'].map(lambda x: genre_list(x))
```

```python
def actor(x):
    act=[]
    st=' '
    for i in x:
        act.append(i.get('name'))
    if act==[]:
        return np.NaN
    else:
        return (act[0])
```

```python
data['actor_1_name']=data['cast'].map(lambda x: actor(x))
```

```python
data['actor_1_name']
```

```python
def actor_2(x):
    act=[]
    for i in x:
        act.append(i.get('name'))
    if act==[] or len(act)<=1:
        return np.NaN
    else:
        return (act[1])
```

```python
data['actor_2_name']=data['cast'].map(lambda x: actor_2(x))
```

```python
data['actor_2_name']
```

```python
def actor_3(x):
    act=[]
    for i in x:
        act.append(i.get('name'))
    if act==[] or len(act)<=2:
        return np.NaN
    else:
        return (act[2])
```

```python
data['actor_3_name']=data['cast'].map(lambda x: actor_3(x))
```

```python
data['actor_3_name']
```

```python
def director(x):
    dt=[];
    st=' '
    for i in x:
        if i.get('job')=='Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']=data['crew'].map(lambda x: director(x))
```
